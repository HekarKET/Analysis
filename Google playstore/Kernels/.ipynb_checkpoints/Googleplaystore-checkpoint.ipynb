```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from pylab import rcParams
%matplotlib inline
```

```python
new_df = pd.read_csv("../Data/googleplaystore.csv")
df = new_df 
```

```python
df = df.dropna(axis=0, how='all')
df
```

### Clean Up

```python
#Deleting row where category is misalabled as 1.9
df.drop(df.loc[df['Category']=='1.9'].index,inplace=True,axis=0)
```

```python
installs = []
for i in df['Installs']:
    install = i.replace(',','')
    install = install.replace('+','')
    installs.append(int(install))
df['Installs'] = installs
```

### Some utility functions

```python
def bar_plot(variable):
    plt.style.use('seaborn')
    var = df[variable]
    varValue = var.value_counts()
    plt.bar(varValue.index, varValue.values)
    plt.xticks(rotation='vertical')
    plt.ylabel('Frequency')
    plt.show()
```

```python
def change(string):
    if 'k' in string:
        string = int(float(string.replace('k','')))
        string *= 1000
        return int(float(string))
    if 'M' in string:
        string = int(float(string.replace('M','')))
        string *= 1000000
        return int(float(string))
    return string
```

#### Which categories have highest installs ?

```python
catbyinstall = df.groupby(["Category"])['Installs'].mean().sort_values(ascending=False)
sns.pointplot(x=catbyinstall.index,y=catbyinstall.values)
plt.xticks(rotation='vertical')
plt.ylabel('Installs in millions')
plt.show()
```

### The family category is most frequent yet it has low performance (Installs wise)
