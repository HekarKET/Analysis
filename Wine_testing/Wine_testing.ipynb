```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline  
```

```python
df = pd.read_csv('datasets_4458_8204_winequality-red.csv')
df.head()
```

```python
ax = sns.countplot(data=df,x='quality')
```

```python
bins = (2, 6, 8)
group_names = [0,1]
df['quality'] = pd.cut(df['quality'], bins = bins, labels = group_names)
```

```python
ax = sns.countplot(data=df,x='quality')
```

## Correlation of every feature with eachother and effect on quality

```python
sns.pairplot(data=df,hue="quality",markers=["o", "s"])
```

## Does Properties like density and pH makes any change in quality?

#### Ploting distrubution

```python
sns.displot(df, x="density",color='c',kde=True,col='quality')
```

```python
sns.displot(df, x="pH",color='r',kde=True,col='quality')
```

#### By seeing the distribution of both properties we can see that they hava almost identical distribution of good quality and bad

```python
sns.set_palette('Pastel1')
fig,ax = plt.subplots(2,2,figsize=(10,10))
plt.subplot(2,2,1)
sns.boxplot(data=df,y = 'density',x='quality')
plt.subplot(2,2,2)
sns.boxplot(data=df,y = 'pH',x='quality')
plt.subplot(2,2,3)
sns.regplot(data=df,x='density',y='quality',logistic=True)
plt.subplot(2,2,4)
sns.regplot(data=df,x='pH',y='quality',logistic=True)
```

#### Even by scatterplot and boxplot we cannot see any classification 

```python
sns.scatterplot(data = df, x="density",y='fixed acidity',hue='quality',palette='BuGn')
```

## Whats the Impact of Total Sulfur Dioxide on Quality?

```python
sns.set_palette('cool')
```

```python
sns.boxplot(data=df,x='quality',y='total sulfur dioxide')
```

```python
sns.displot(df, x="total sulfur dioxide",color='y',kde=True,col='quality')
```

```python
sns.barplot(data=df,x='quality',y='total sulfur dioxide')
```

### Even though it is undetectable in wine still it makes a good classification on wine

## Checking Remaining Categories and their Impact on Quality Of Wine

```python
sns.set_palette('RdBu')
```

```python
cat = ['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','sulphates','alcohol']
fig,ax = plt.subplots(3,3,figsize=(15,10))

for i,y in enumerate(cat,1):
    plt.subplot(3,3,i)
    sns.barplot(data=df,x='quality',y=y)
```

### Citric acid and Volatile acidity makes very big differnce in quality
